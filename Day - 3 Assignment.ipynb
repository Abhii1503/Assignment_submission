1. Use Condtional Statements to write a program of Report Card.

```python
#Marks to Grade Calculator

m = input("Enter the Marks to Calculate the Grade:\t")
m = int(m)
if  m>=95:
    print("Grade : A+\n")
elif 85<=m<95:
    print("Grade : A\n")
elif 75<=m<85:
    print("Grade : B+\n")
elif 65<=m<75:
    print("Grade : B\n")
elif 55<=m<65:
    print("Grade : C+\n")
elif 45<=m<55:
    print("Grade : C\n")
elif 35<=m<45:
    print("Grade : D+")
elif 25<=m<35:
    print("Grade : D")
else:
    print("Fail")
```

2. Print All prime numbers Between 1 to 1000

```python
for i in range(1,1001,1):
    
    #skipping 0 and 1 as they are neither prime nor composite
    if(i==0 or i==1):
        continue;
    
    #flag variable will tell whether the no. is prime or not
    flag = 1;
    
    for j in range(2,((i//2)+1),1):
        if(i%j==0):
            flag=0;
            break;
    
    # flag = 1 means No. is prime
    # flag = 0 means No. is Not Prime
    if(flag==1):
        print(i)
```

3.Printing tables from 1 to 10 

```python
for i in range(1,11):
    print("\n")
    for j in range(1,11):
        print(i*j)
```

4. Program to Print X Prime Number using While Loop where X is input From user

```python
maximum = int(input(" Please Enter the Maximum Value: "))

Number = 0

while(Number <= maximum):
    count = 0
    i = 2
    
    while(i <= Number//2):
        if(Number % i == 0):
            count = count + 1
            break
        i = i + 1

    if (count == 0 and Number != 1):
        print(" %d" %Number, end = '  ')
    Number = Number  + 1
```
